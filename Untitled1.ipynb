```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds

#https://www.tensorflow.org/tutorials/load_data/images
```

```python
print(tf.__version__)
```

```python
import pathlib
data_dir = pathlib.Path('planets')
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

```python
roses = list(data_dir.glob('Earth/*'))
PIL.Image.open(str(roses[100]))
```

```python
batch_size = 32
img_height = 256
img_width = 256
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

# Preprocessing

## Normalizing

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
```

```python
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model('modelP')
```

```python
model.summary()
```

```python
model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])
```

```python
model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=3
)

```

```python
loss, accuracy = model.evaluate(val_ds)
print("Validation accuracy:", accuracy)
print("Validation loss:", loss)
```

```python
import os
import shutil
import numpy as np
from imblearn.over_sampling import SMOTE
from imblearn import pipeline
from imblearn.pipeline import make_pipeline
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

# Path to your original dataset
data_dir = pathlib.Path('planets')

# Path to store the oversampled dataset
oversampled_dir = pathlib.Path('oversampled_planets')

# Desired number of samples per class
desired_samples = 150

# Create a SMOTE pipeline
oversample = SMOTE(sampling_strategy='auto', k_neighbors=2)
pipeline = make_pipeline(oversample)

# Iterate over each class (planet)
for planet in planets:
    # Path to the class folder
    planet_dir = data_dir / planet

    # List all images in the class folder
    images = list(planet_dir.glob('*'))

    # Calculate the number of times to duplicate each image
    duplication_factor = desired_samples // len(images)

    # Duplicate images
    for image in images:
        for i in range(duplication_factor):
            shutil.copy(image, oversampled_dir / planet)

    # Copy the remaining images to reach the desired number of samples
    remaining_samples = desired_samples % len(images)
    for i in range(remaining_samples):
        shutil.copy(images[i], oversampled_dir / planet)

    # Apply SMOTE to further oversample the class
    X = np.array(images)
    X = X.reshape(-1, 1)  # Reshape for SMOTE
    X_resampled, _ = pipeline.fit_resample(X, X)
    X_resampled = X_resampled.reshape(-1)  # Reshape back to original shape

    # Copy the resampled images to the oversampled dataset folder
    for i, image in enumerate(X_resampled):
        shutil.copy(images[i % len(images)], oversampled_dir / planet)
```
